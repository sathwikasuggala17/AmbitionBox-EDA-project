### problem statement : Conducting an analysis on all the leading startup's companies operating in India

### importing libraries

```python
#importing necessary libraries for EDA 
import pandas as pd
import numpy as np
import regex as re
import seaborn as sns
import matplotlib.pyplot as plt

# core libraries to collect the data
# importing requests, along with library
import requests
from bs4 import BeautifulSoup
```

#### step-2 data collection
- Extracting startup details from ambitionbox website using web scraping techniques.
#### fetching the data

```python
# Initialize lists to store data
companies = []
industry = []
employee_rate = []
startup_type = []
establishment_year = []
company_locations = []
company_rating = []
reviews = []
salaries = []
interviews = []
jobs = []
benefits = []

# Loop through the pages
for i in range(1, 81):  # Adjusted loop range
    url = f'https://www.ambitionbox.com/list-of-companies?sortBy=popular&tags=startup&page={i}'

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3',
        'Accept-Language': 'en-US,en;q=0.5',
    }

    response = requests.get(url, headers=headers)

# max_retries = 3  # Adjust as needed
# retry_delay = 5  # Adjust as needed

# for i in range(1, 81):  # Adjusted loop range
#     url = f'https://www.ambitionbox.com/list-of-companies?sortBy=popular&tags=startup&page={i}'

#     headers = {
#         'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3',
#         'Accept-Language': 'en-US,en;q=0.5',
#     }

#     for attempt in range(max_retries):
#         try:
#             response = requests.get(url, headers=headers)
#             response.raise_for_status()
#             break  # Break out of the retry loop if successful
#         except requests.exceptions.RequestException as e:
#             if attempt < max_retries - 1:
#                 print(f"Retrying after {retry_delay} seconds due to connection error: {e}")
#                 time.sleep(retry_delay)
#             else:
#                 raise  # Raise the exception if all retries are exhausted
    
    soup = BeautifulSoup(response.text, "html.parser")
    container = soup.find_all('div', class_='companyCardWrapper')
    
    # Loop through the containers on the page
    for i in container:
        # companies column -1
        company_name_element = i.find('h2', class_="companyCardWrapper__companyName")
        company_name = company_name_element.text.strip()
        companies.append(company_name if company_name else 'np.nan')
            
        # industry column -2
        s = i.find('span', class_="companyCardWrapper__interLinking")
        if s:
            q = s.text.strip()
        match = re.findall(r'^([^|\d]+)', q)
        if match:
            extracted_text = match[0].strip()
           
            industry.append(extracted_text)
        else:
            industry.append('np.nan')
            
        # employee_rate column -3
        er = i.find('span', class_="companyCardWrapper__interLinking")
        employees = re.findall('\d+k\-?\d+?k?\sEmployees', er.text)
        if len(employees)>0:
            employee_rate.append(employees[0])
        else:
            employee_rate.append(np.nan)
            
    # startup_type column -3
        st = i.find('span', class_="companyCardWrapper__interLinking")
        _type = re.findall('Employees\s\|\s(.*)\|\s\d', st.text)
        if len(_type)>0:
            startup_type.append(_type[0])
        else:
            startup_type.append(np.nan)
            
    # establishment_year column -4
        ey = i.find('span', class_="companyCardWrapper__interLinking")
        if ey:
            q = ey.text.strip()
        match = re.findall(r'(\d+)\s*years\s*old', q)
        if match:
            extracted_text = match[0].strip()
            establishment_year.append(extracted_text)
        else:
            establishment_year.append('np.nan')
            
    # company_locations column -5
        cl =i.find('span', class_="companyCardWrapper__interLinking")
        locations = re.findall('\d+\syears\sold\s\|(.*)', cl.text)
        if len(_type)>0:
            company_locations.append(locations[0])
        else:
            company_locations.append(np.nan)
            
    # company_rating column- 6
    for i in soup.find_all('div', class_='companyCardWrapper'):
        cr_value = i.find_all('span', class_="companyCardWrapper__companyRatingValue")
        for i in cr_value:
            company_rating.append(i.text)

    # reviews column-7
    for i in soup.find_all('div',class_="companyCardWrapper__tertiaryInformation"):
        revws = re.findall('(.*)Reviews', i.text)
        if len(revws)>0:
            reviews.append(''.join(revws))
        else:
            reviews.append(np.nan)
            
    # salaries column = 8
    for i in soup.find_all('div', class_="companyCardWrapper__tertiaryInformation"):
        salary_info = re.findall('Reviews(.*)Salaries', i.text)
        if len(salary_info) > 0:
            salaries.append(''.join(salary_info))
        else:
            salaries.append(np.nan)
        
     # interviews column - 9
    for i in soup.find_all('div', class_="companyCardWrapper__tertiaryInformation"):
        interview_info = re.findall('Salaries(.*)Interviews', i.text)
        if len(interview_info) > 0:
            interviews.append(''.join(interview_info))
        else:
            interviews.append(np.nan)
            
    # jobs column - 10
    for i in soup.find_all('div', class_="companyCardWrapper__tertiaryInformation"):
        jobs_info = re.findall('Interviews(.*)Jobs', i.text)
        if len(jobs_info) > 0:
            jobs.append(''.join(jobs_info))
        else:
            jobs.append(np.nan)

    # benefits column - 11
    for i in soup.find_all('div', class_="companyCardWrapper__tertiaryInformation"):
        benefits_info = re.findall('Jobs(.*)\sBenefits', i.text)
        if benefits_info:
            benefits.append(''.join(benefits_info))
        else:
            benefits.append(np.nan)

print(len(companies), len(industry), len(startup_type), len(establishment_year), len(company_locations), len(company_rating), len(reviews), len(salaries), len(interviews), len(jobs), len(benefits))
```

```python
startup_details_new ={'companies' : companies,
                      'industry' : industry,
                      'startup_type' : startup_type,
                      'establishment_year' : establishment_year,
                      'company_locations' : company_locations,
                      'company_rating' : company_rating,
                      'reviews' : reviews,
                      'salaries' : salaries,
                      'interviews' : interviews,
                      'jobs' : jobs,
                      'benefits' : benefits}
```
